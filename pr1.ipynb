```python
num1=int(input("Enter first number:"))
num2=int(input("Enter second number:"))
print()

add=num1+num2
print("addition:",add)
print()

sub=num1-num2
print("subtraction:",sub)
print()

mul=num1*num2
print("multiplication:",mul)
print()

div=num1/num2
print("division:",div)
print()

square=num1**2
print("square of first number is:",square)
```

```python
txt1 = input("Enter first string: ")
txt2 = input("Enter second string: ")
print()

concatenation = txt1 + txt2
print("Concatenation:", concatenation)
print()

print("To upper case:", txt1.upper())
print("To lower case:", txt2.lower())
print()

print("Length of first string:", len(txt1))
print("Length of second string:", len(txt2))
print()

print("Capitalize:", txt1.capitalize())
print()

print("Center:", txt1.center(6, '*'))
print()

print("Left Justify:", txt1.ljust(10, '-'))
print()

print("Right Justify:", txt2.rjust(10, '-'))
print()

old = input("Enter the word/character to replace from first string: ")
new = input("Enter the new word/character: ")
print("After Replace:", txt1.replace(old, new))
print()

print("Before:", txt2)
print("After:", txt2.strip())
```
